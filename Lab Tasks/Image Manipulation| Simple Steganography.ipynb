## Helper Functions

```python
def binary_to_str(binary):
    try:
        int(len(binary) / 8)
    except Exception as e:
        raise Exception('Binary length is not multiple of 8')

    string = ''
    print(len(binary) // 8)
    for i in range(len(binary) // 8):
        integer = int(''.join(binary[8 * i:8 * (i + 1)]), 2)
        ascii_character = chr(integer)
        string += ascii_character

    return string
```

```python
def decode_lsb(image_path):
    #Read the image in RGB and get the first channel value
    img = cv2.imread(image_path,cv2.IMREAD_COLOR)[:,:,0]
    # Iterate over the image row by row, get the pixel value and make a list of strings (zeros and ones)
    img_list = [str(x) for x in img.ravel()%2]
    #Convertion to string
    message = binary_to_str(img_list)
    return message
```

```python
img_file = 'secret-happy-dog.png'
decode_lsb(img_file)
```
